## Program

```python
sigma = np.array([[14.0, 8.0, 3.0],
              [8.0, 5.0, 2.0],
              [3.0, 2.0, 1.0]])
BB = np.array([[1.0, 2.0, 3.0],
              [0.0, 1.0, 2.0],
              [0.0, 0.0, 1.0]])

U, D, V = np.linalg.svd(BB)
sig_inv = np.linalg.inv(sigma) 
D = np.diag(D)
print(U)
print(D)
print(V)
u1 = U.T[0].T
print(u1)
x = np.array([1.052, 0.6648, 0.2271])
z1 = u1 * (u1.T @ x.T)
print(z1)
print(u1.T @ z1)
```

## Reproduce Figure 4.12

```python
N = 10000
dim = 3
xx = np.array(np.random.randn(N,dim))
for i in range(N) :
  while xx[i][0]*xx[i][0]+xx[i][1]*xx[i][1]+xx[i][2]*xx[i][2] > 1 :
    xx[i][0] = np.random.randn()
    xx[i][1] = np.random.randn()
    xx[i][2] = np.random.randn()

xx = np.reshape(xx, (N,dim))
print(xx)
fig = plt.figure()
ax = plt.axes(projection='3d')

#for i in range(N):
#    ax.scatter(xx[i][0], xx[i][1], xx[i][2])
ax.scatter3D(xx.T[0], xx.T[1], xx.T[2])

plt.show()
```
