## Example 3.2
## Import modules

```python
""" bvnormal.py """
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
print(np.__version__)
```

## Set random seed

```python
np.random.seed(12)
```

## Reproduce Figure 3.1 (left)

```python
N = 100000
r = 0.0 #change to 0.8 for other plot
Sigma = np.array([[1, r], [r, 1]])
B = np.linalg.cholesky(Sigma)
x = B @ randn(2,N)
plt.scatter([x[0,:]],[x[1,:]], alpha =0.4, s = 4)
plt.show()
```

## Reproduce Figure 2.13 (left)

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

hist, xedges, yedges = np.histogram2d(x[0,:],x[1,:], bins=40, range=[[-3, 3], [-3, 3]])

# The start of each bucket.
xpos, ypos = np.meshgrid(xedges[:-1], yedges[:-1])

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)

# The width of each bucket.
dx, dy = np.meshgrid(xedges[1:] - xedges[:-1], yedges[1:] - yedges[:-1])

dx = dx.flatten()
dy = dy.flatten()
dz = hist.flatten()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```

## Reproduce Figure 3.1 (right)

```python
r = 0.8 #change to 0.8 for other plot
Sigma = np.array([[1, r], [r, 1]])
B = np.linalg.cholesky(Sigma)
x = B @ randn(2,N)
plt.scatter([x[0,:]],[x[1,:]], alpha =0.4, s = 4)
plt.show()
```
